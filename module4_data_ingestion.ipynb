# Module 4: Data Handling\n\n## Data Ingestion Notebook Using Pandas\n\nThis notebook loads CSV, JSON, and text data, cleans the datasets, performs simple exploratory data analysis, and exports a cleaned CSV file.

## 1. Import Required Libraries

```python
from pathlib import Path\nimport pandas as pd
```

## 2. Set File Paths

```python
DATA_DIR = Path('data')\nOUTPUT_DIR = Path('output')\nOUTPUT_DIR.mkdir(exist_ok=True)\n\ncsv_path = DATA_DIR / 'patients.csv'\njson_path = DATA_DIR / 'diagnoses.json'\ntext_path = DATA_DIR / 'patient_comments.txt'\ncleaned_output_path = OUTPUT_DIR / 'cleaned_patient_data.csv'
```

## 3. Load CSV Data\n\nThe CSV file contains basic patient visit information.

```python
patients_df = pd.read_csv(csv_path)\npatients_df.head()
```

## 4. Load JSON Data\n\nThe JSON file contains diagnosis and risk-level information.

```python
diagnoses_df = pd.read_json(json_path)\ndiagnoses_df.head()
```

## 5. Load Text Data\n\nThe text file uses a pipe separator, so it can still be loaded with `pd.read_csv()`.

```python
comments_df = pd.read_csv(text_path, sep='|')\ncomments_df.head()
```

## 6. Normalize Column Names\n\nThis makes column names easier to work with by converting them to lowercase and replacing spaces with underscores.

```python
def normalize_columns(df):\n    df = df.copy()\n    df.columns = (\n        df.columns\n        .str.strip()\n        .str.lower()\n        .str.replace(' ', '_')\n        .str.replace('-', '_')\n    )\n    return df\n\npatients_df = normalize_columns(patients_df)\ndiagnoses_df = normalize_columns(diagnoses_df)\ncomments_df = normalize_columns(comments_df)\n\npatients_df.head()
```

## 7. Inspect Missing Values

```python
patients_df.isnull().sum()
```

## 8. Clean the Patient Dataset\n\nCleaning steps:\n\n- Fill missing age with the median age\n- Fill missing city with `Unknown`\n- Fill missing total cost with the median cost\n- Fill missing follow-up value with `No`\n- Remove duplicate rows

```python
patients_clean = patients_df.copy()\n\npatients_clean['age'] = patients_clean['age'].fillna(patients_clean['age'].median())\npatients_clean['city'] = patients_clean['city'].fillna('Unknown')\npatients_clean['total_cost'] = patients_clean['total_cost'].fillna(patients_clean['total_cost'].median())\npatients_clean['follow_up_needed'] = patients_clean['follow_up_needed'].fillna('No')\n\npatients_clean = patients_clean.drop_duplicates()\n\npatients_clean
```

## 9. Clean the Diagnosis Dataset

```python
diagnoses_clean = diagnoses_df.copy()\ndiagnoses_clean['diagnosis'] = diagnoses_clean['diagnosis'].fillna('Unknown')\ndiagnoses_clean['notes'] = diagnoses_clean['notes'].fillna('No notes provided')\n\ndiagnoses_clean.head()
```

## 10. Simple Exploratory Data Analysis

```python
patients_clean.head()
```

```python
patients_clean.describe()
```

```python
patients_clean['visit_type'].value_counts()
```

```python
patients_clean['city'].value_counts()
```

```python
diagnoses_clean['risk_level'].value_counts()
```

## 11. Combine the Datasets\n\nThe datasets are joined using `patient_id`.

```python
combined_df = patients_clean.merge(diagnoses_clean, on='patient_id', how='left')\ncombined_df = combined_df.merge(comments_df, on='patient_id', how='left')\n\ncombined_df.head()
```

## 12. Final Missing Value Check

```python
combined_df.isnull().sum()
```

## 13. Export the Cleaned Dataset

```python
combined_df.to_csv(cleaned_output_path, index=False)\nprint(f'Cleaned CSV exported to: {cleaned_output_path}')
```

## 14. Conclusion\n\nThis notebook loaded CSV, JSON, and text data using pandas. It cleaned missing values, normalized column names, removed duplicate rows, performed simple exploratory data analysis, combined the datasets, and exported a cleaned CSV file.
